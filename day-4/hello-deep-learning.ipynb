## Hello Deep Learning
In the next cell you are going to execute the 'hello world' version of deep learning: The MNIST dataset. Please run the next cell. It is no necessary to understand what's going on in the code. We will learn over the course :) 

```
import tensorflow as tf

# 1. Load data
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()

# 2. Preprocess data
x_train, x_test = x_train / 255.0, x_test / 255.0

# 3. Define mode's architecture
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

sgd = tf.keras.optimizers.SGD(lr=0.01)
model.compile(optimizer=sgd,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train model
model.fit(x_train, y_train, epochs=1, validation_split=0.2)

# 5. Evaluate model
score = model.evaluate(x_test, y_test)
print('Test accuracy:', score[1])
```
